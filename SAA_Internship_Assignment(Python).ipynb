Q1 Write a program that computes the net amount of a bank account based a transaction log from
console input. The transaction log format is shown as following:
D 100
W 200
D means deposit while W means withdrawal.
Suppose the following input is supplied to the program:
D 300
D 300
W 200
D 100
Then, the output should be:
500
Hints:
In case of input data being supplied to the question, it should be assumed to be a console input.

```python
inp = input('Enter Deposit or Withdrawal Amount with "D" or "W" as prefix')
D = int()
W = int()
spl = inp.split()
num = 0
mapped_transaction = []
try:
    for i in range(len(spl)):
        mapped_transaction.append((spl[num],spl[num+1]))
        num+=2
except:
    pass

for transaction in mapped_transaction:
    if transaction[0] == 'D':
        D += int(transaction[1])
    if transaction[0] == 'W':
        W += int(transaction[1])
        
print(f"The Closing Balance is {D-W}")
```

### Q19 Please write a program using generator to print the numbers which can be divisible by 5 and 7 between 0 and n in comma separated form while n is input by console.

```python
def five_seven_divisor(n):
    for divident in range(1, n+1):
        if divident % 5 == 0 and divident % 7 ==0:
            yield divident

# function call to give it a user input argument
div = five_seven_divisor(int(input('enter a natural number: ')))

# iterating through the generator, to get all the values in csv form
lst = []
for num in div:
    lst.append(str(num))
print(','.join(lst))
```
